**1. Most popular neighbourhoods (with respect to booked nights)**

```python
df2['booked_nights'] = 365 - df['Unoccupied days\n(in last 365 days)']
df2['booked_nights']
```

```python
popular_neighbourhoods = df2.groupby('neighbourhood')['booked_nights'].sum().sort_values(ascending=False)
popular_neighbourhoods
```

```python

plt.figure(figsize=(10, 6))
popular_neighbourhoods.head(10).plot(kind='bar', color='lightblue')
plt.title('Top 10 Most Popular Neighbourhoods by Booked Nights')
plt.xlabel('Neighbourhood')
plt.ylabel('Total Booked Nights')
plt.xticks(rotation=90)
plt.show()
```

De Baarsjes - Oud-West leads in Airbnb bookings with over 1.1 million nights, closely followed by De Pijp - Rivierenbuurt at approximately 801,000 nights.

Centrum-West and Centrum-Oost are also popular, recording 638,094 and 498,782 booked nights, respectively.

Westerpark and Zuid show strong booking figures with 471,779 and 430,709 nights, respectively.

The visualization highlights significant disparities in booking popularity, with neighborhoods like Bijlmer-Centrum and Bijlmer-Oost showing much lower figures.

**2. Highest revenue with respect to neighbourhoods**

```python
df2['Total_Revenue']= df2['price per night'] * df2['booked_nights']

revenue_by_neighbourhood= df2.groupby('neighbourhood')['Total_Revenue'].sum().sort_values(ascending= False)

revenue_by_neighbourhood
```

```python
plt.figure(figsize=(10, 6))
revenue_by_neighbourhood.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Neighbourhoods by Revenue')
plt.xlabel('Neighbourhood')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.show()
```

 De Baarsjes - Oud-West generated revenue over 157 million, followed by De Pijp - Rivierenbuurt at 120 million.

Centrum-West and Centrum-Oost are also earned well, recording 115 million and 89 million

The visualization highlights significant disparities in revenue generated by top 10 neighborhoods.

**3. Best overall listings- (Weightage - 0.6 value of booking, 0.4 number of reviews)**

```python
df2['listing_score'] = 0.6 * df2['Total_Revenue'] + 0.4 * df2['number_of_reviews']
best_listings = df2[['id', 'name', 'listing_score']].sort_values(by='listing_score', ascending=False)
best_listings
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x=best_listings['name'].head(10), height=best_listings['listing_score'].head(10), color='pink')
plt.title('Top 10 Best Overall Listings')
plt.xlabel('Listing Name')
plt.ylabel('Score')
plt.xticks(rotation=90)
plt.show()
```

Property Ranking: The graph reveals the ranking of properties based on their overall listing scores. The property "Luxurious, Beautiful & Unique appt." has the highest score, significantly higher than the others, indicating it as the top-ranked listing.

Score Distribution: There is a large gap between the top-ranked property and the next few properties, suggesting a significant difference in their overall listing scores. However, the scores for the next four properties are relatively close, indicating a tighter competition among them.

Property Categories: The listings can be broadly categorized into two distinct groups based on their scores. The top five properties form the high-scoring group, while the remaining five properties form the lower-scoring group, with a noticeable gap between the two groups.


**4. Perform Correlation Analysis : Find out which factors causing more booked nights

```python
# Calculate the correlation matrix
correlation_matrix = df2[['booked_nights','price per night', 'number_of_reviews', 'reviews_per_month','Total_Revenue','listing_score', 'Host Listings Count']].corr()

# Display the correlation matrix
correlation_matrix
```

```python
sns.set(style='whitegrid')

# Create a correlation matrix
plt.figure(figsize=(10, 6), facecolor='white')
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Variables')
plt.show()
```

The variables that show a notable correlation with the number of booked nights include:

price per night: Negative correlation (-0.1698)

number of reviews: Negative correlation (-0.2348)

reviews per month: Negative correlation (-0.2267)

Total Revenue: Positive correlation (0.2808)

listing score: Positive correlation (0.2806)

These correlations provide insights into the factors that may influence the number of booked nights for listings.

The number of booked nights has a moderate positive correlation of 0.28 with both total revenue and listing score, indicating that listings with more booked nights tend to generate higher revenue and have better listing score

     5.2 Booking by reviews (bucketing)

```python
# Bucketing reviews into categories
df2['review_category'] = pd.cut(df2['number_of_reviews'], bins=[0, 10, 50, 100, 500, 1000, float('Inf')], labels=['0-10', '11-50', '51-100', '101-500', '501-1000', '>1000'])
review_performance = df2.groupby('review_category')['booked_nights'].sum()
review_performance
```

```python
review_performance.plot(kind='bar', color='purple')
plt.title('Booking by Review Categories')
plt.xlabel('Review Categories')
plt.ylabel('Total Booked Nights')
plt.xticks(rotation=0)
plt.show()
```

The Bar Graph represents the relation between Review and Booking.

The overwhelming majority of bookings fall into the 0-10 bin, accounting for approximately 51.9% (2,856,209 bookings) of the total 5,474,821 bookings.

The 11-50 bin represents the second-largest segment, with 2,124,731 bookings or around 38.8% of the total.

The 51-100 bin accounts for a relatively small portion of bookings at 298,188 or 5.4% of the total.

The 101-500 bin represents 3.5% (191,675 bookings) of the total bookings, indicating a smaller but non-negligible segment.

The 501-1000 bin has the smallest number of bookings at 4,118, comprising only 0.08% of the total bookings.


**6. Sorted list of business done by all hosts (descending order)**

```python
business_by_hosts = df2.groupby('host_id')['Total_Revenue'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
business_by_hosts.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Hosts by Total Revenue')
plt.xlabel('Host ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.show()
```

There is a significant disparity in revenue generated among the top 10 hosts, with a steep dropoff after the first two hosts.

The top host appears to be generating an outstanding revenue of around $10 million, far surpassing the other hosts.

The second-highest revenue generator is bringing in approximately $5.1 million, which is still substantially lower than the top host's revenue.

The third and fourth highest revenue generators are earning around $2.5 million  and  $1.9 million, respectively, representing a notable gap from the top two hosts.

The remaining six hosts in the top 10 are generating revenues ranging from approximately  $1.8 million down to around $1 million.

The distribution of revenue among the top 10 hosts appears to follow the Pareto principle or the 80/20 rule, where a small percentage of hosts (the top 2) are responsible for a large portion of the total revenue.
    
This concentrated revenue distribution could potentially indicate the presence of extremely successful or high-performing hosts, as well as opportunities for other hosts to optimize their strategies and offerings to increase their revenue generation.
